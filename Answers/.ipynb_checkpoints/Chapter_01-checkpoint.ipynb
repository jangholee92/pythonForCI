```python
# 1. Sum of all even numbers from 1 to 10,000
even_sum = sum(range(2, 10001, 2))
print("Sum of all even numbers from 1 to 10,000:", even_sum)
```

```python
# 2. Generate a random array and calculate mean and standard deviation
import numpy as np
random_array = np.random.rand(100, 100)
mean_value = np.mean(random_array)
std_value = np.std(random_array)
print("Mean:", mean_value)
print("Standard Deviation:", std_value)
```

```python
# 3. Compound interest calculation
P = 1000  # Principal
r = 0.05  # Annual interest rate
t = 10    # Time in years
A = P * (1 + r)**t
print("Compound interest after 10 years: $", round(A, 2))
```

```python
# 4. Find numbers divisible by both 3 and 7 from 1 to 1,000
divisible = [x for x in range(1, 1001) if x % 3 == 0 and x % 7 == 0]
print("Numbers divisible by both 3 and 7:", divisible)
```

```python
# 5. Compute the given expression
import math
result = (15 * 12) / (3**2 + 8) + math.log(100) - math.sqrt(64)
print("Result of the expression:", round(result, 2))
```

```python
# 6. Generate random temperatures and calculate statistics
temps = np.random.uniform(-30, 50, 365)
hottest_day = np.max(temps)
coldest_day = np.min(temps)
avg_temp = np.mean(temps)
print("Hottest day temperature:", round(hottest_day, 2))
print("Coldest day temperature:", round(coldest_day, 2))
print("Average temperature:", round(avg_temp, 2))
```

```python
# 8. Compute distance between two points
x1, y1 = 2, 3  # Point 1
x2, y2 = 5, 7  # Point 2
distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
print("Distance between points:", round(distance, 2))
```

```python
# 9. Find second largest and second smallest numbers in a list
numbers = [4, 7, 1, 9, 3, 8]
unique_numbers = list(set(numbers))
unique_numbers.sort()
second_largest = unique_numbers[-2]
second_smallest = unique_numbers[1]
print("Second largest:", second_largest)
print("Second smallest:", second_smallest)
```

```python
# 10. Calculate the median of a list
numbers = [5, 2, 9, 1, 7]
median_value = np.median(numbers)
print("Median:", median_value)
```

```python
# 11. Compute sum of squares of first n positive integers
n = 10**6
loop_sum = sum([i**2 for i in range(1, n+1)])
formula_sum = n * (n + 1) * (2 * n + 1) // 6
print("Sum of squares (loop):", loop_sum)
print("Sum of squares (formula):", formula_sum)
```

```python
# 12. Simulate rolling dice
num_trials = 10000
count = 0
for _ in range(num_trials):
    dice_1 = np.random.randint(1, 7)
    dice_2 = np.random.randint(1, 7)
    if dice_1 + dice_2 in [7, 11]:
        count += 1
probability = count / num_trials
print("Probability of rolling a 7 or 11:", probability)
```

```python
# 13. Daily temperature averages and fluctuations
temps = np.random.uniform(0, 35, (7, 24))  # 7 days, 24 hours
daily_avg = np.mean(temps, axis=1)
daily_fluctuations = np.ptp(temps, axis=1)
day_with_max_fluctuation = np.argmax(daily_fluctuations)
print("Daily averages:", daily_avg)
print("Day with max fluctuation:", day_with_max_fluctuation + 1)
```

```python
# 14. Check if two strings are anagrams
def are_anagrams(str1, str2):
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()
    return sorted(str1) == sorted(str2)
result = are_anagrams("Listen", "Silent")
print("Are the strings anagrams?", result)
```

```python
# 15. Generate Fibonacci sequence until first term exceeds 10,000
fibonacci = [0, 1]
while True:
    next_term = fibonacci[-1] + fibonacci[-2]
    if next_term > 10000:
        break
    fibonacci.append(next_term)
print("Fibonacci sequence:", fibonacci)
```

```python
# 16. Replace temperatures above 50°C with 50
temps = np.array([45, 52, 30, 55, 48])
clamped_temps = np.where(temps > 50, 50, temps)
print("Clamped temperatures:", clamped_temps)
```

```python
# 17. Calculate wind chill index
def calculate_wind_chill(temp, wind_speed):
    return 13.12 + 0.6215 * temp - 11.37 * wind_speed**0.16 + 0.3965 * temp * wind_speed**0.16
wind_chill = calculate_wind_chill(5, 15)
print("Wind Chill Index:", round(wind_chill, 2))
```

```python
# 18. Find outliers in data
data = [10, 12, 15, 18, 100, 14, 9, 8, 110]
mean = np.mean(data)
std = np.std(data)
outliers = [x for x in data if abs(x - mean) > 2 * std]
print("Outliers:", outliers)
```

```python
# 19. ClimateData class
class ClimateData:
    def __init__(self, temperatures, rainfall):
        self.temperatures = temperatures
        self.rainfall = rainfall

    def avg_temperature(self):
        return np.mean(self.temperatures)

    def total_rainfall(self):
        return np.sum(self.rainfall)

    def wettest_day(self):
        return np.max(self.rainfall)

climate = ClimateData([20, 22, 18, 25], [12, 15, 20, 10])
print("Average Temperature:", climate.avg_temperature())
print("Total Rainfall:", climate.total_rainfall())
print("Wettest Day Rainfall:", climate.wettest_day())
```

```python
# 20. Identify months exceeding average temperature
temps = [5, 7, 10, 15, 18, 22, 25, 24, 20, 14, 10, 6]
avg_temp = np.mean(temps)
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
above_avg_months = [months[i] for i in range(12) if temps[i] > avg_temp]
print("Months above average temperature:", above_avg_months)
```
