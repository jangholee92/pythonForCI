```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Ensure plots display inline
%matplotlib inline
```

```python
# 1. Simulate daily temperatures for one year and calculate monthly averages
dates = pd.date_range(start="2023-01-01", end="2023-12-31")
temps = np.random.uniform(-10, 40, len(dates))
df_temps = pd.DataFrame({'Date': dates, 'Temperature': temps})
df_temps['Month'] = df_temps['Date'].dt.month
monthly_avg = df_temps.groupby('Month')['Temperature'].mean()

plt.figure(figsize=(10, 6))
monthly_avg.plot(kind='bar', color='skyblue')
plt.title("Monthly Average Temperature")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.grid(axis='y')
plt.show()
```

```python
# 2. Rainfall heatmap for cities and months
rainfall = np.random.randint(0, 300, (10, 10))
cities = [f"City {i+1}" for i in range(10)]
months = [f"Month {i+1}" for i in range(10)]
df_rainfall = pd.DataFrame(rainfall, index=cities, columns=months)

plt.figure(figsize=(8, 6))
plt.imshow(df_rainfall, cmap='Blues', aspect='auto')
plt.colorbar(label="Rainfall (mm)")
plt.title("Rainfall Heatmap")
plt.xticks(ticks=np.arange(10), labels=months, rotation=45)
plt.yticks(ticks=np.arange(10), labels=cities)
plt.show()
```

```python
# 3. Hourly temperatures and 6-hour averages
timestamps = pd.date_range(start="2023-01-01 00:00", periods=48, freq="H")
hourly_temps = np.random.uniform(15, 30, len(timestamps))
df_hourly = pd.DataFrame({'Timestamp': timestamps, 'Temperature': hourly_temps})
df_hourly.set_index('Timestamp', inplace=True)
six_hour_avg = df_hourly.resample('6H').mean()

plt.figure(figsize=(10, 6))
plt.plot(df_hourly.index, df_hourly['Temperature'], label="Hourly Temperatures")
plt.plot(six_hour_avg.index, six_hour_avg['Temperature'], label="6-Hour Average", linestyle='--', color='red')
plt.title("Hourly Temperatures and 6-Hour Averages")
plt.xlabel("Time")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid()
plt.show()
```

```python
# 4. CO₂ emissions by country and year
emissions = np.random.randint(100, 1000, (12, 5))
countries = [f"Country {i+1}" for i in range(12)]
years = [f"Year {i+1}" for i in range(5)]
df_emissions = pd.DataFrame(emissions, index=countries, columns=years)
df_emissions['Total'] = df_emissions.sum(axis=1)

plt.figure(figsize=(12, 6))
df_emissions.drop('Total', axis=1).plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title("CO₂ Emissions by Country (Grouped by Year)")
plt.xlabel("Country")
plt.ylabel("Emissions (tons)")
plt.legend(title="Year")
plt.show()
```

```python
# 5. Daily sales for 6 products
dates = pd.date_range(start="2023-01-01", periods=30)
products = [f"Product {i+1}" for i in range(6)]
sales = np.random.randint(50, 200, (30, 6))
df_sales = pd.DataFrame(sales, index=dates, columns=products)
df_sales_monthly = df_sales.sum()

plt.figure(figsize=(10, 6))
df_sales_monthly.plot(kind='bar', color='purple')
plt.title("Total Monthly Sales per Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
# 6. High and low temperatures
temps_high = np.random.uniform(10, 35, 30)
temps_low = np.random.uniform(-5, 10, 30)
dates = pd.date_range(start="2023-01-01", periods=30)
df_temps = pd.DataFrame({'Date': dates, 'High Temp': temps_high, 'Low Temp': temps_low})
df_temps['Range'] = df_temps['High Temp'] - df_temps['Low Temp']
max_range_day = df_temps.loc[df_temps['Range'].idxmax()]

plt.figure(figsize=(12, 6))
plt.plot(df_temps['Date'], df_temps['High Temp'], label="High Temp")
plt.plot(df_temps['Date'], df_temps['Low Temp'], label="Low Temp")
plt.fill_between(df_temps['Date'], df_temps['High Temp'], df_temps['Low Temp'], color='lightblue', alpha=0.5)
plt.title("Daily High and Low Temperatures")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.show()
```

```python
# 7. Sensor readings and hourly averages
timestamps = pd.date_range(start="2023-01-01 00:00", periods=1000, freq="T")
sensor_readings = np.random.uniform(50, 150, len(timestamps))
df_sensors = pd.DataFrame({'Timestamp': timestamps, 'Reading': sensor_readings})
df_sensors.set_index('Timestamp', inplace=True)
hourly_avg = df_sensors.resample('H').mean()

plt.figure(figsize=(10, 6))
hourly_avg.plot(kind='bar', color='green')
plt.title("Hourly Average Sensor Readings")
plt.xlabel("Hour")
plt.ylabel("Reading")
plt.show()
```

```python
# 8. Monthly rainfall totals over 10 years
years = list(range(2013, 2023))
months = list(range(1, 13))
rainfall = np.random.randint(50, 300, (10, 12))
df_rainfall = pd.DataFrame(rainfall, index=years, columns=months)
annual_totals = df_rainfall.sum(axis=1)

plt.figure(figsize=(10, 6))
annual_totals.plot(kind='line', marker='o', color='orange')
plt.title("Yearly Total Rainfall")
plt.xlabel("Year")
plt.ylabel("Rainfall (mm)")
plt.grid()
plt.show()
```

```python
# 9. Scatter plot of temperature vs. humidity with rainfall as point size
dates = pd.date_range(start="2023-01-01", periods=365)
temperature = np.random.uniform(-10, 40, 365)
humidity = np.random.uniform(30, 90, 365)
rainfall = np.random.uniform(0, 50, 365)
df_weather = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity, 'Rainfall': rainfall})

plt.figure(figsize=(10, 6))
plt.scatter(df_weather['Temperature'], df_weather['Humidity'], s=df_weather['Rainfall']*2, alpha=0.6, c='blue')
plt.title("Temperature vs. Humidity (Rainfall as Size)")
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")
plt.grid()
plt.show()
```

```python
# 10. Hourly energy consumption for one week and daily totals
timestamps = pd.date_range(start="2023-01-01", periods=24*7, freq='H')
energy_usage = np.random.uniform(0.5, 2.5, len(timestamps))
df_energy = pd.DataFrame({'Timestamp': timestamps, 'Energy (kWh)': energy_usage})
df_energy.set_index('Timestamp', inplace=True)
daily_totals = df_energy.resample('D').sum()

plt.figure(figsize=(10, 6))
daily_totals.plot(kind='bar', color='purple')
plt.title("Daily Energy Consumption")
plt.xlabel("Day")
plt.ylabel("Total Energy (kWh)")
plt.grid(axis='y')
plt.show()
```

```python
# 11. Stock prices and percentage change
dates = pd.date_range(start="2023-01-01", periods=30)
companies = ['Company A', 'Company B', 'Company C', 'Company D', 'Company E']
stock_prices = np.random.uniform(100, 200, (30, 5))
df_stocks = pd.DataFrame(stock_prices, index=dates, columns=companies)
df_pct_change = df_stocks.pct_change() * 100

plt.figure(figsize=(12, 6))
for company in companies:
    plt.plot(df_pct_change.index, df_pct_change[company], label=company)

plt.title("Percentage Change in Stock Prices")
plt.xlabel("Date")
plt.ylabel("Percentage Change (%)")
plt.legend()
plt.grid()
plt.show()
```

```python
# 12. Cumulative rainfall step plot
timestamps = pd.date_range(start="2023-01-01", periods=500, freq='D')
rainfall = np.random.uniform(0, 50, len(timestamps))
df_rainfall = pd.DataFrame({'Date': timestamps, 'Rainfall (mm)': rainfall})
df_rainfall.set_index('Date', inplace=True)
cumulative_rainfall = df_rainfall.cumsum()

plt.figure(figsize=(10, 6))
plt.step(cumulative_rainfall.index, cumulative_rainfall['Rainfall (mm)'], where='mid', color='blue', label='Cumulative Rainfall')
plt.title("Cumulative Rainfall Over Time")
plt.xlabel("Date")
plt.ylabel("Rainfall (mm)")
plt.grid()
plt.legend()
plt.show()
```

```python
# 13. Monthly average temperatures for two cities
dates = pd.date_range(start="2023-01-01", end="2023-12-31")
temps_city1 = np.random.uniform(-5, 35, len(dates))
temps_city2 = np.random.uniform(-5, 35, len(dates))
df_temps = pd.DataFrame({'Date': dates, 'City 1': temps_city1, 'City 2': temps_city2})
df_temps.set_index('Date', inplace=True)
df_temps['Month'] = df_temps.index.month
monthly_avg = df_temps.groupby('Month').mean()

plt.figure(figsize=(10, 6))
plt.plot(monthly_avg.index, monthly_avg['City 1'], label="City 1", marker='o')
plt.plot(monthly_avg.index, monthly_avg['City 2'], label="City 2", marker='s')
plt.title("Monthly Average Temperatures for Two Cities")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid()
plt.show()
```

```python
# 14. Multi-axis plot: temperature, wind speed, precipitation
timestamps = pd.date_range(start="2023-01-01", periods=24, freq='H')
temperature = np.random.uniform(10, 30, 24)
wind_speed = np.random.uniform(0, 15, 24)
precipitation = np.random.uniform(0, 10, 24)

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot temperature on primary y-axis
ax1.plot(timestamps, temperature, color='red', label="Temperature (°C)")
ax1.set_xlabel("Time")
ax1.set_ylabel("Temperature (°C)", color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Plot wind speed and precipitation on secondary y-axis
ax2 = ax1.twinx()
ax2.plot(timestamps, wind_speed, color='blue', linestyle='--', label="Wind Speed (km/h)")
ax2.bar(timestamps, precipitation, color='green', alpha=0.3, label="Precipitation (mm)")
ax2.set_ylabel("Wind Speed (km/h) / Precipitation (mm)", color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

fig.suptitle("Weather Data: Temperature, Wind Speed, and Precipitation")
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.95))
fig.tight_layout()
plt.show()
```

```python
# 15. Best-selling product by month
dates = pd.date_range(start="2023-01-01", periods=365)
products = [f"Product {i+1}" for i in range(10)]
sales = np.random.randint(10, 100, (365, 10))
df_sales = pd.DataFrame(sales, index=dates, columns=products)
df_sales['Month'] = df_sales.index.month
monthly_totals = df_sales.groupby('Month').sum()

best_selling = monthly_totals.idxmax(axis=1)

plt.figure(figsize=(10, 6))
monthly_totals.plot(kind='bar', stacked=True, colormap='Paired')
plt.title("Monthly Sales for Best-Selling Products")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(axis='y')
plt.show()
```

```python
# 16. Daily maximum AQI with threshold line
dates = pd.date_range(start="2023-01-01", periods=7)
hourly_aqi = np.random.uniform(0, 500, (7, 24))  # AQI readings for a week (7 days, 24 hours)
df_aqi = pd.DataFrame(hourly_aqi, index=dates, columns=[f"Hour {i}" for i in range(1, 25)])
daily_max_aqi = df_aqi.max(axis=1)

plt.figure(figsize=(10, 6))
plt.plot(daily_max_aqi.index, daily_max_aqi, marker='o', label="Daily Max AQI", color='purple')
plt.axhline(100, color='red', linestyle='--', label="Threshold AQI = 100")
plt.title("Daily Maximum AQI")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.legend()
plt.grid()
plt.show()
```

```python
# 17. Temperature anomalies heatmap with extreme value annotations
years = list(range(2013, 2025))
months = list(range(1, 13))
anomalies = np.random.uniform(-3, 3, (len(years), len(months)))  # Temperature anomalies
df_anomalies = pd.DataFrame(anomalies, index=years, columns=months)

plt.figure(figsize=(12, 6))
plt.imshow(df_anomalies, cmap='coolwarm', aspect='auto')
plt.colorbar(label="Temperature Anomaly (°C)")
plt.title("Temperature Anomalies Heatmap")
plt.xticks(ticks=np.arange(12), labels=months, rotation=45)
plt.yticks(ticks=np.arange(len(years)), labels=years)

# Annotate extreme values
for i in range(len(years)):
    for j in range(12):
        if abs(df_anomalies.iloc[i, j]) > 2.5:
            plt.text(j, i, f"{df_anomalies.iloc[i, j]:.1f}", ha='center', va='center', color='black')
plt.show()
```

```python
# 18. Scatter plot of temperature vs. rainfall for three cities
dates = pd.date_range(start="2023-01-01", periods=365)
city1_temps = np.random.uniform(-5, 35, 365)
city1_rainfall = np.random.uniform(0, 50, 365)
city2_temps = np.random.uniform(-5, 35, 365)
city2_rainfall = np.random.uniform(0, 50, 365)
city3_temps = np.random.uniform(-5, 35, 365)
city3_rainfall = np.random.uniform(0, 50, 365)

plt.figure(figsize=(12, 8))
plt.scatter(city1_temps, city1_rainfall, label="City 1", alpha=0.6)
plt.scatter(city2_temps, city2_rainfall, label="City 2", alpha=0.6)
plt.scatter(city3_temps, city3_rainfall, label="City 3", alpha=0.6)
plt.title("Temperature vs. Rainfall")
plt.xlabel("Temperature (°C)")
plt.ylabel("Rainfall (mm)")
plt.legend()
plt.grid()
plt.show()
```

```python
# 19. Daily sales and peak hour averages
dates = pd.date_range(start="2023-01-01", periods=7, freq='D')
hourly_sales = np.random.randint(50, 200, (7, 24))  # Hourly sales for a week
df_sales = pd.DataFrame(hourly_sales, index=dates, columns=[f"Hour {i}" for i in range(24)])
peak_hours = df_sales.iloc[:, 12:18]  # 12 PM to 6 PM
daily_avg_peak_sales = peak_hours.mean(axis=1)

plt.figure(figsize=(10, 6))
plt.bar(daily_avg_peak_sales.index, daily_avg_peak_sales, color='orange')
plt.title("Average Peak Hour Sales (12 PM - 6 PM)")
plt.xlabel("Date")
plt.ylabel("Average Sales")
plt.grid(axis='y')
plt.show()
```

```python
# 20. Coldest and hottest weeks of the year
dates = pd.date_range(start="2023-01-01", periods=365)
daily_temps = np.random.uniform(-10, 40, 365)
df_temps = pd.DataFrame({'Date': dates, 'Temperature': daily_temps})
df_temps['Week'] = df_temps['Date'].dt.isocalendar().week
weekly_avg = df_temps.groupby('Week')['Temperature'].mean()
coldest_week = weekly_avg.idxmin()
hottest_week = weekly_avg.idxmax()

coldest_week_temps = df_temps[df_temps['Week'] == coldest_week]
hottest_week_temps = df_temps[df_temps['Week'] == hottest_week]

plt.figure(figsize=(12, 6))
plt.plot(coldest_week_temps['Date'], coldest_week_temps['Temperature'], label=f"Coldest Week (Week {coldest_week})", color='blue')
plt.plot(hottest_week_temps['Date'], hottest_week_temps['Temperature'], label=f"Hottest Week (Week {hottest_week})", color='red')
plt.title("Coldest and Hottest Weeks of the Year")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid()
plt.show()
```
